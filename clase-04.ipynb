### ¬øEs primo el numero?

```python
# SOLID -> Single responsability
# Crear una funci√≥n llamada "es_primo" que reciba un n√∫mero positivo como par√°metro -> devolver True / False
```

```python
def es_primo(n):
    n_entero = int(n)
    # Solo puede dividirse por 1 y por si mismo
    # El n√∫mero 1 no es primo
    
    # deber√≠a lanzar una exception
    # early return
    if n_entero < 0:
        return "Solo aceptamos numeros positivos o cero"
    if n_entero <= 1:
        return False
    for i in range(2, n_entero):
        if n_entero % i == 0: # si esto sucede, tiene divisores => NO es primo
            return False
    # si llegamos hasta ac√°
    return True
    
    #¬†la unica condici√≥n que queda es que sea mayor a cero
    
```

```python
es_primo(2)
```

```python
# hacer una funci√≥n que clasifique el riesgo de una persona del titanic
# Si es clase 1 y menor o = a 15 a√±os -> bajo
# Si es clase 3 y tiene m√°s de 60 a√±os -> riesgo alto
# Sino (else) riesgo medio
```

```python
df.iloc[0]
```

```python
def clasificar_riesgo(pasajero):
    print(pasajero)
    edad = pasajero["Age"]
    clase = pasajero["Pclass"]
    
    if edad <= 15 and clase == 1:
        return "bajo"
    if edad > 60 and clase == 3:
        return "alto"
    return "medio"
```
